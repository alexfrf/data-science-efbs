## Objetivo
El objetivo de estas sesiones es proporcionar herramientas de preparación de datos que nos permitan extraer la máxima información de nuestros datasets.

Estas técnicas nos permiten tanto limpiar los datos existentes como incorporar nuevos atributos que no existían originalmente o descartar aquellos que no aportan información útil.

Desde el máster reunimos las técnicas de Feature Engineering en dos grandes bloques:

* Limpieza y transformación
* Modificación de la dimensionalidad (reducción/incremento)

## Limpieza y Transformación
La limpieza de datos incluye todo el conjunto de procesos que deben ser realizados para poder sacar conclusiones válidas sobre los datos.

Este tipo de tratamiento van encaminados a evitar que nuestro análisis quede sesgado por un número de observaciones muy pequeño o por inconsistencias de información.

En este apartado se incluyen también las transformaciones que debemos realizar para que la información pueda ser consumidosa por determinados algoritmos (especialmente normalización y estandarización).

En este grupo de técnicas se incluye:

* Conversión de tipos de datos
* Relleno de información faltante (tratamiento de nulos)
* Sustitución de valores anómalos (tratamiento de outliers)
* Escalado de la distribución (normalización, estandarización)
* Codificación de variables

## Modificación de la dimensionalidad

Entendemos por modificación de la dimensionalidad todos los procesos cuyo objetivo sea añadir o eliminar columnas del juego de datos original para facilitar su análisis o mejorar la calidad de los datos.

### Reducción de la dimensionalidad
La reducción de la dimensionalidad es un apartado clave al trabajar con grandes conjuntos de datos. 

El tiempo que tarde un algoritmo de analítica avanzada depende de dos factores:

- La complejidad del propio algoritmo
- El tamaño de la entrada

La complejidad del algoritmo es un factor intrínseco que no podemos cambiar, hay métodos más rápidos y otros más lentos y, aunque podemos ayudar a que sean más rápidos (proporcionando los datos con unos formatos específicos), no podemos acelerar algunas medidas elementales de un algoritmo como son el número de operaciones que tiene que hacer por iteración.

El tamaño de la entrada, sin embargo, sí podemos modificarlo. Consideramos tamaño de la entrada a la cardinalidad total del conjunto de datos con el que trabajamos, es decir, nos importa tanto el número de filas como de columnas.

Si pensamos en un conjunto de datos real en el que tengamos información de venta, clientes, proveedores, producto, centro de venta, comercial que realiza la venta, ... podemos ver que el número de características que debemos tener en cuenta es muy elevado.

Los procesos de reducción de la dimensionalidad se encargan de determinar qué atributos/características tienen más importancia de cara a determinar el resultado de nuestros procesos de analítica avanzada y nos permiten eliminar aquellos que no tengan aporte con la confianza de no estar condicionando nuestros análisis.

Las técnicas habituales para la reducción de la dimensionalidad son:
* Selección de características manual
* Análisis de correlación (y eliminación de variables que tengan una muy alta correlación)
* PCA
* LASSO

### Incremento de la dimensionalidad

El incremento de la dimensionalidad no es una técnica (o conjunto de técnicas) reconocido como tal, normalmente nos referimos a este paso como "enriquecimiento de los datos" o nombres similares.

En el máster, este paso se incluye aquí por su paralelismo con la reducción de la dimensionalidad del apartado anterior.

En general las técnicas son más manuales que las de la reducción de la dimensionalidad:
* Desagregación de atributos "empaquetados"
* Etiquetado de casos (etiquetado por personas, no por algoritmos)
* Introducción de datos externos

En estos "datos externos" también se incluiría el uso de las salidas de otros modelos analíticos como ampliación de características para nuestro juego de datos, por ejemplo, utilizar la clasificación de productos de un modelo de scoring como entrada para decidir la efectividad de la oferta por categoría de producto.

# Limpieza y Transformación

Una vez tenemos un entendimiento básico de los datos (suele ser conveniente hacer un pequeño documento resumiendo lo que hemos encontrado), se trata de comenzar con las tareas de limpieza de datos.

En este paso queremos terminar con un conjunto de datos que sea fácilmente utilizable.

Algunas tareas a realizar en este paso serán:
* Eliminar atributos que no utilizaremos ("Unnamed: 0" en el ejemplo, otros ejemplos serían índices, valores de verificación, fechas de integración de los datos, códigos de verificación,...)
* Corregir tipos de datos ("pub_date" en el ejemplo es un object aunque contiene una fecha)
* Imputar valores a los nulos (no existe casi ningún nulo en el ejemplo a excepción de author_type que tiene aproximadamente 4000 nulos)
* Elimimar información redundante (en el ejemplo pub_day, pub_month, pub_year y pub_weekday pueden obtenerse de pub_date), ¡ojo!, en función de los algoritmos que necesitemos aplicar puede ser beneficioso tener este tipo de información redundante.



## Eliminar atributos que no utilizaremos

Aunque más adelante se verá el proceso de reducción de la dimensionalidad, existe determinada información que puede ser eliminada de forma directa, en el caso de ```pitchfork``` esta serída la columna *Unnamed: 0*.

* Tenemos un atributo numérico y su equivalente descriptivo (eliminamos el descriptivo, aunque habitualmente, mantendremos el mapeo en una variable por si fuese preciso.

## Tratar los duplicados

Tras eliminar los atributos que no nos interesen suele ser una buena práctica tratar los valores duplicados.

El motivo de realizar este paso después de la eliminación de atributos es que, podemos tener observaciones que estén duplicadas únicamente después de eliminar algunos atributos (por ejemplo, si tenemos ventas a nivel producto-día-hora-minuto-segundo y eliminamos hora-minuto-segundo, podemos tener duplicados a nivel producto-día)

Por lo tanto, hay conjuntos de datos en los que es legítimo tener duplicados en la entrada, como es el caso anterior en el que podríamos tener una tabla como la siguiente:

|CLIENTE | PRODUCTO | FECHA | IMPORTE |
| -- | -- | -- | -- |

### Agregación de datos

En los casos en que tiene sentido que tengamos líneas duplicadas lo que debemos hacer es agregar esas líneas en una sola a nivel **clave primaria**.

El concepto de clave primaria se estudió en bases de datos y, resumiendo, es el menor conjunto de datos que identifica, de forma única, una fila de la tabla.

En un DataSet, también tenemos que tener ese concepto de clave primaria ya que, si tenemos atributos duplicados, los algoritmos que se verán en el resto del máster no funcionarán bien.

Para eliminar duplicados mediante agregación solo tenemos que hacer un groupby por clave primaria aplicando la operación adecuada en cada caso para quedanros con un único registro a nivel clave primaria (suma para ventas, promedio y suma para clientes presentes en una cafetería,...)


````python
df = pd.read_csv('./data/1. fact_venta.csv', sep=';', decimal=',')
````

### Eliminación de duplicados

Otro método para tratar los duplicados es, simplemente, borrarlos, para ello debemos estar seguros de que los datos que vamos a borrar no son necesarios para nuestro análisis y que son realmente registros duplicados.

Si trabajamos con datos de Pitchfork, no tendría sentido que existan duplicados ya que, sin entrar mucho en cómo organice Pitchfork su trabajo, suponemos que cada **reviewid** debería ser único.

Por lo tanto, cualquier duplicado será un error de procesado previo y debe ser eliminado.

En un caso de datos reales obtenidos de sistemas transaccionales de una empresa, no sería suficiente con eliminar los duplicados sino que habría que entender cómo y por qué se han generado por si pueden estar corrompiendo otros datos.

En pandas, la eliminación de duplicados es muy sencilla:

> ```df.drop_duplicates()```

````python
def leer_pitchfork():
    df = pd.read_csv('./data/reviews.csv')
    df.drop(columns=['Unnamed: 0'], inplace=True)
    df = df.drop_duplicates()
    return df

df_pitch = leer_pitchfork()
df_pitch = df_pitch.drop_duplicates()
````

Una vez hemos realizado las tareas básicas de lectura y limpieza de datos suele ser una buena idea ir creando funciones que nos permitan replicar el trabajo realizado hasta el momento de forma rápida.

En este caso crearemos una función ```leer_pitchfork``` que podemos utilizar en el resto del cuaderno para que realice la tarea de lectura, eliminado de columnas innecesarias y de duplicados y que nos devuelva el dataframe listo para ser utilizado.

## Evaluación de los nulos

Al igual que en la base de datos diferenciamos dos tipos de nulos en una base de datos; atributos que son nulos pero que deberían tener valor y atributos que son nulos y de los que nos es imposible conocer su valor o no lo tienen.

Por ejemplo, un pedido de una venta online realizado pero no entregado no tendrá fecha de entrega, en este caso no debemos tratar de cubrir ese dato sino que es correcto que esté en blanco.

Desde el punto de vista de los algoritmos de analítica avanzada, un nulo solamente debería estar en campos para los que queremos predecir el valor.

Savlo que un atributo esté en esa situación, debemos siempre proporcionar valor a todos los campos o eliminarlos del modelo.

Sin embargo, cuando los datos deberían tener un valor podemos proceder a su imputación, siguiendo diferentes estrategias según el caso:

* Cubrir con 0: si el dato no existe, por ejemplo, en un negocio no hay venta un día, debemos rellenarlo con ceros.
* Imputar valores: Si sabemos que debe existir un valor pero no tenemos forma de obtenerlo por ningún medio podemos imputar un valor que no altere las propiedades del atributo:
    * Si es un atributo cuantitativo podemos imputar la media.
    * Si es un atributo cualitativo podemos imputar "Otros", "Desconocido" o "NCOP" (no contenido en otras partes)
* Interpolar: si tenemos valor anterior y valor siguiente y los datos deben pasar, forzosamente por los estados intermedios, podemos rellenar con el valor interpolado (la temperatura debe pasar por estados intermedios, el stock en una tienda,...)
* Fill-forward, fill-backward: en algunas ocasiones puede ser interesante rellenar los valores faltantes propagando hacia delante o hacia atrás el anterior/siguiente valor que se haya registrado.


Puesto que el Dataframe de Pitchfork no tiene valores cuantitativos nulos, utilizaremos el de COVID para probar la imputación de valores numéricos.

### Interpolar

La interpolación ya se había visto en un caso anterior y, simplemente, calcula la curva entre dos puntos de acuerdo a diferentes métodos:

````python
# Si no decimos nada se hace la interpolación lineal

df_covid = pd.read_csv('./data/owid-covid-data.csv')
df_esp = df_covid.loc[df_covid.iso_code == 'ESP',:]
df_esp.icu_patients.interpolate().plot()
````

## Tipos de datos

Desde un punto de vista analítico, tendríamos la siguiente taxonomía de datos:

* Cuantitativos (métricas):
    * Continuos (un valor continuo, por ejemplo temperatura)
    * Discretos (un valor exacto, por ejemplo número de visitantes de una tienda)
* Cualitativos (dimensionales):
    * Categóricos (diferentes valores sin orden, por ejemplo tipos de electrodomésticos)
        * Dicotómicos (caso particular que solamente puede tomar dos valores si-no, encima-debajo...)
    * Ordinales (diferentes valores pero que sí guardan un orden, por ejemplo escala Likert: muy bien, bien, normal, mal, muy mal)

Más allá de la tipología de datos, está el formato de datos que utiliza un lenguaje de programación para representarlos. 

A este formato se le llama también tipo de datos. Pandas infiere, para cada celda de datos un tipo (type) con el que almacena el valor en memoria.

Desde el punto de vista de pandas hay varios tipos de datos diferentes:

| Tipo | Explicación | Tipo de datos en python |
|:---:|:---|---|
| int64 | Números enteros | int |
| float64 | Números con decimales | float |
| bool | Valores cierto o falso | bool |
| datetime64 | Fechas y horas (precisión hasta segundos) | datetime |
| timedelta[ns] | Sellos horarios (precisión hasta nanosegundos) | -- |
| object | Bolsa mixta de otros tipos de datos o datos tipo texto | str |
| category | Atributo categórico | -- |

Por desgracia Pandas no siempre acierta con el tipo de datos a asignar y, en muchas ocasiones, es necesario realizar conversiones de tipos para que la información que almacenamos esté en el formato necesario para su análisis.

Parte del trabajo inicial en un dataset consistirá en asegurar que el formato de representación de los datos (el que usa pandas) es apto para su tipología.

Para ello realizaremos unas sencillas conversiones:
* Todos los datos que sean de tipo categórico, independientemente de que sean números o texto, deberían convertirse a texto (tenemos datos categóricos numéricos cuando, por ejemplo, tenemos el código de producto y no su descriptivo)

### Convertir a datetime

Pandas, al igual que SQL, admite dos tipos diferentes de conversiones de tipos, aquellas que pandas considera directas y las que requieren un intento "más en serio".

En SQL para este cometido hay dos funciones: 
* CAST => le dice a SQL que trate un tipo como si fuese otro, por ejemplo '3' en lugar de 3
* CONVERT => le dice a SQL que intente convertir un tipo de datos en otro, por ejemplo '20220121' en 21/01/2022

En Pandas el equivalente sería 
* astype => cambia la representación para conversiones simples
* to_xxx => convierte el tipo de dato

Pandas tiene tres métodos to_xxx:
* ```to_numeric```: convertir texto en número (ya sea con decimales o sin ellos)
* ```to_datetime```: convertir texto en datetime
* ```to_timedelta```: convertir texto en timedelta

En este caso tenemos que utilizar ```to_datetime``` para la conversión:


````python
pd.to_datetime(df_pitch['pub_date'])
````

````python
# Aunque Pandas puede inferir el formato de una columna fecha, por seguridad en la conversión y velocidad, 
# es conveniente especificar el formato en el que viene el campo.
# %Y => Año en formato 4 dígitos
# %m => Mes en formato 2 dígitos
# %d => día en formato 2 dígitos
# El listado completo de códigos puede consultarse en: 
#      https://docs.python.org/3/library/datetime.html#strftime-strptime-behavior

df_pitch.pub_date = pd.to_datetime(df_pitch.pub_date, format='%Y-%m-%d')

df_pitch.dtypes
````

## Ejercicio 5. Función Benford

Escribe una función que acepte como parámetros una serie de datos numéricos y una "tolerancia" y devuelva como resultado si la serie de datos pasa el test de Benford

````python
df = pd.read_excel('./data/precios_venta.xlsx')
````

````python
def check_benford(df,muestra):
    df = df[df['MUESTRA'].str.contains(str(muestra))]
    df['PRIMER DIGITO'] = df['PRECIO'].astype(str).str[0]
    gr = round((df['PRIMER DIGITO'].value_counts() / df['PRIMER DIGITO'].value_counts().sum())*100,2)
    idx = list(gr.index)
    lst_ = []
    j = 0
    for i in idx:
        i = int(i)
        k = i-j
        lst_.append(k)
        j = i
    if list(np.ones((9))) == lst_:
        print('Se cumple la Ley de Benford para la muestra {}'.format(muestra).upper())
    else:
        print('No se cumple la Ley de Benford para la muestra {}'.format(muestra).upper())
    print(gr)

for i in range(1,5+1):
    check_benford(df,i)
    print('\n')
````
