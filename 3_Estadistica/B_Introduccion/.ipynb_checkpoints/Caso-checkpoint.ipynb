```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from statsmodels.formula.api import ols
```

```python
df = pd.read_excel('DS-MDS002-Locos por el Futbol.xlsx')
df['DEL_S'] = np.where(df['POS']=='DEL',1,0)
df['Rank1_S'] = np.where(df['RankClub']==1,1,0)
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse=False)
for i in ['PIE','RankFiFa','RankClub']:
    codif = encoder.fit_transform(df[[i]])
    df[encoder.get_feature_names()] = codif

df.columns
```

```python
jug = df.copy()
```

```python
sns.histplot(df['PT_GBP'],bins=10)
```

```python
from sklearn.preprocessing import PowerTransformer
p = PowerTransformer(method='yeo-johnson', standardize=True)

transformada = p.fit_transform(df[['PT_GBP']] )
df['PT_GBP_t'] = transformada

sns.histplot(df['PT_GBP_t'], bins=10)
```

```python
df.describe()
```

```python
media = df['PT_GBP_t'].mean()
std = df['PT_GBP_t'].std()
k = 2*std
df['Outlier'] = 0
df.loc[abs(df['PT_GBP_t']-media) > k, 'Outlier'] = 1
```

```python
outliers = df[df['Outlier']==1]
outliers
```

```python
df1 = df.copy()
df = df[df['Outlier']!=1]
df.drop('Outlier',axis=1,inplace=True)
```

```python
df = df[['Nombre Jugador','Desde','A','PT_GBP','ALT','EdadTraspaso','%GolTit','%T_N2','x0_Z','x0_A','Rank1_S','PT_GBP_t','DEL_S']]
jug=df.copy()
df.drop(['Nombre Jugador','Desde','A'],axis=1,inplace=True)
df.columns = ['PT_GBP','ALT','EdadTraspaso','GolTit','T_N2','x0_Z','x0_A','Rank1_S','PT_GBP_t','DEL_S']
k = ['ALT','EdadTraspaso','GolTit','T_N2','x0_Z','x0_A','Rank1_S']
```

```python
X = df.drop('PT_GBP',axis=1)
X = X[['GolTit','Rank1_S']]
y = df['PT_GBP']
lr = LinearRegression()
lr.fit(X, y)
print(lr.score(X,y))

a = lr.intercept_
b = lr.coef_
```

```python
reg = ols('PT_GBP_t ~ Rank1_S+GolTit', data = df).fit(cov_type='HC3') # Assuming homoscedasticity
print(reg.summary())
```

```python
def get_y(df=df):
    y_pred = lr.predict(X)
    df['Valor_Real'] = y_pred
    df['Valor_Real'] = round(df['Valor_Real'])
    return df


jug = get_y(jug)
jug['DIF'] = jug['Valor_Real'] - jug['PT_GBP']
```

```python
jug['A'] = jug['A'].str.strip()
jug['Desde'] = jug['Desde'].str.strip()
jug['A'] = jug['A'].str.replace('AS Monaco','Monaco')
jug['Desde'] = jug['Desde'].str.replace('AS Monaco','Monaco')
```

```python
desde = jug.groupby('Desde',as_index=False)['DIF'].agg(['sum','count']).reset_index()
a = jug.groupby('A',as_index=False)['DIF'].agg(['sum','count']).reset_index()
```

```python
k = pd.merge(desde,a,how='outer',left_on='Desde',right_on='A').fillna(0)
k['Club'] = np.where(k['Desde']!=0,k['Desde'],k['A'])
k['PlusVenta'] = k['sum_x']*-1
k['PlusCompra'] = k['sum_y']
k['count'] = k['count_x'] + k['count_y']
k= k[['Club','count','PlusVenta','PlusCompra']]
```

```python
k['DIF'] = k['PlusVenta'] + k['PlusCompra']
k = k.sort_values(by='DIF',ascending=False)
```
