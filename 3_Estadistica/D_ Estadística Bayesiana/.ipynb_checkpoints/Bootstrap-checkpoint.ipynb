# Bootstrap

Bootstrap es un método de remuestro que se basa en tratar la muestra que hemos recogido en un experimento como si se tratase de la población total.

Es importante conocer qué ofrece Bootstrap:
* Permite estimar la incertidumbre
* No mejora la medida obtenida de forma inicial

Podemos pensar en bootstrap como una simulación de qué hubiese pasado si, en lugar de conseguir exactamente nuestras encuestas/mediciones/resultados hubiésemos tenido unos un poco diferentes.


Con esta suposición lo que realiza son remuestreos del mismo tamaño que la muestra original pero permitiendo repetición de elementos.

Imaginemos que hemos realizado mediciones de la temperatura de una piscina y obtenemos las siguientes medidas:

```python
from matplotlib import pyplot

pyplot.rcParams['figure.figsize'] = (16,12)

medidas = [16, 16.2, 16.1, 15.4, 17]

media_original = sum(medidas) / len(medidas)

print('media original:', media_original)
```

Bootstrap realizaría **m** muestras sobre el original de 5 elementos cada una; normalmente **m** es un número elevado.

```python
from random import choice

nuevas_medidas = []
for k in range(len(medidas)):
    nuevas_medidas.append(choice(medidas))
    
nuevas_medidas

print('nueva media:', sum(nuevas_medidas) / len(nuevas_medidas))
```

```python
medias = []

for w in range(int(1E6)):
    nuevas_medidas = []
    for k in range(len(medidas)):
        nuevas_medidas.append(choice(medidas))
    medias.append(sum(nuevas_medidas) / len(nuevas_medidas))
    
```

```python
import seaborn as sns

sns.histplot(medias)

print('media global:', sum(medias)/ len(medias))
```

Vemos que la media obtenida mediante Bootstrap es casi igual a la media original pero, la distribución de probabilidades ahora está centrada en la media y admite desviaciones según los posibles valores que puede tomar la muestra.

Como nuestra muestra original era muy pequeña no pueden formarse todos los valores que formarían el continuo entre 15.4 y 17.0.

Si le damos más medidas obtenemos un resultado más suave:

```python
medidas = [15.4, 15.5, 15.6, 15.7, 15.8, 15.9, 16, 16.1, 16.2, 16.3,16.4,16.5,16.6,16.7]
print('media original:', media_original)
medias = []

for w in range(int(1E6)):
    nuevas_medidas = []
    for k in range(len(medidas)):
        nuevas_medidas.append(choice(medidas))
    medias.append(sum(nuevas_medidas) / len(nuevas_medidas))
    
```

```python
sns.histplot(medias)

print('media global:', sum(medias)/ len(medias))
```

Por supuesto, python dispone de funciones que ayudan a conseguir el mismo resultado escribiendo menos código:

```python
from sklearn.utils import resample
import numpy as np

medidas = [15.4, 15.5, 15.6, 15.7, 15.8, 15.9, 16, 16.1, 16.2, 16.3,16.4,16.5,16.6,16.7]

len_m = len(medidas)

medias = []
for w in range(int(1E6)):
    x = resample(medidas)
    medias.append(sum(x) / len_m)
    
```

```python
sns.histplot(medias)

print('media global:', sum(medidas) / len(medidas))
```

## Cálculo intervalo de confianza

Como hemos visto en los aprtados anteriores, la distribución que genera Bootstrap con estos datos de entrada es casi normal.

Ahí podríamos aplicar los intervalos de confianza conocidos (media + x desviaciones típicas).

Ojo, no tendría por qué serlo, podemos meter valores extremos y ver cómo varía el resultado:


```python
from sklearn.utils import resample
import numpy as np

medidas = [11.4, 15.5, 15.6, 15.7, 15.8, 15.9, 16, 16.1, 16.2, 16.3,16.4,16.5,16.6,23.7]

len_m = len(medidas)

medias = []
for w in range(int(1E6)):
    x = resample(medidas)
    medias.append(sum(x) / len_m)
    
sns.histplot(medias)

print('media global:', sum(medidas) / len(medidas))    
```

Una vez tenemos nuestra distribución generada con bootstrap es posible calcular sus intervalos de confianza por cualquier método de los que hemos estudiado, si no se encuentra una relació con nignuna distribución conocida, podemos utilizar el método del rango intercuartiles.
