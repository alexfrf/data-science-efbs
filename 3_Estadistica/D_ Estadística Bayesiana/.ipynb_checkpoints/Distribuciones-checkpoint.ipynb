# Distribuciones estadísticas

## Objetivo

Tener una pequeña guía de cómo generar determinadas distribuciones estadísticas con Python y algunas de sus características.

## Por qué es de interés
Tenemos dos usos principales de las distribuciones estadísticas en ciencia de datos:

* Contraste y validación de hipótesis (sesión anterior)
* Modelado estadístico

El modelado estadístico consiste en utilizar valores extraídos de una distribución para entender cómo cambia la respuesta de nuestro sistema ante diferentes situaciones.

El uso principal del modelado estadístico es la simulación.


## Uso en Python
Tenemos dos librerías que nos ayudan a trabajar con distribuciones estadísticas:

* numpy, módulo random
* scipy, módulo stats

**Scipy** trabaja con la distribución completa mientras que **numpy** solamente extrae valores acordes a la distribución.

Su uso, salvo situaciones específicas, depende de nuestras preferencias.

### scipy.stats

scipy.stats utiliza el método rvs para extraer valores de la distribución

### numpy.random 
numpy.random utiliza como método el propio nombre de la distribución, cada distribución tiene parámetros diferentes.

## Binomial

Muestra el número de éxitos esperables en un proceso aleatorio con determinado número de intentos.

### Uso

Número de piezas fallidas, número de piezas rechazadas...

```python
from scipy.stats import binom

# número de intentos
# probabilidad de éxito
# estado aleatorio
# tamaño
data = binom.rvs(10,.1,1, 1000)

sns.histplot(data=data)

print(data.min(), data.max(), data.mean(), data.std())
```

```python
from numpy import random as rn
data = rn.binomial(10, .1, size=1000)

sns.histplot(data=data)

print(data.min(), data.max(), data.mean(), data.std())
```

```python
from numpy import random as rn
data = rn.binomial(10, .5, size=1000)

sns.histplot(data=data)

print(data.min(), data.max(), data.mean(), data.std())
```

## Poisson

Simula el número de eventos que se producen a lo largo del tiempo.

### Uso

Personas que llegan a una parada de autobús, colas de supermercado...

```python
from scipy.stats import poisson

# número de intentos
# probabilidad de éxito
# estado aleatorio
# tamaño
data = poisson.rvs(.5,0, 1000)

sns.histplot(data=data)

print(data.min(), data.max(), data.mean(), data.std())
```

```python
from numpy import random as rn
data = rn.poisson(.5, 1000)

sns.histplot(data=data)

print(data.min(), data.max(), data.mean(), data.std())
```

```python
from numpy import random as rn
data = rn.poisson(5, 1000)

sns.histplot(data=data)

print(data.min(), data.max(), data.mean(), data.std())
```

# Continuas aproximadas

## Uniforme

La distribución uniforme no tiene ninguna complicación, se trata de una distribución que devuelve cualquier valor con la misma probabilidad.

### Uso

Utilizamos una distribución uniforme si queremos representar sucesos equiprobables o si solamente conocemos el rango de valores que puede tomar una distribución.

```python
from scipy.stats import uniform

# 10 es la media
# 1 es el rango
# 1000 es el número de muestras a extraer
data = uniform.rvs(10,1,1000)
sns.histplot(data=data)

print(data.min(), data.max(), data.mean(), data.std())
```

```python
from numpy import random as rn

data = rn.uniform(10,11, 1000)

sns.histplot(data=data)

print(data.min(), data.max(), data.mean(), data.std())
```

## Triangular

Simula la forma de un triángulo.

### Uso

Conocemos el rango de los valores y el valor más probable.

```python
from scipy.stats import triang

# media
# intentos
data = triang.rvs(0.40, size=1000)

sns.histplot(data=data)

print(data.min(), data.max(), data.mean(), data.std())
```

```python
from numpy import random as rn
data = rn.triangular(0, 25,100, 1000)

sns.histplot(data=data)

print(data.min(), data.max(), data.mean(), data.std())
```

## Continuas

### Beta
Simula tiempos para completar una tarea, proporciones, requiere poca información para poder generarse.

Beta acepta dos parámetros que actúan como si fuesen dos mitades de una distribución normal.

```python
from scipy.stats import beta

# media
# intentos
data = beta.rvs(5,10, size=10000)

sns.histplot(data=data, bins=100)

print(data.min(), data.max(), data.mean(), data.std())
```

```python
from numpy import random as rn

data = rn.beta(10,1, size=10000)
sns.histplot(data=data, bins=100)

print(data.min(), data.max(), data.mean(), data.std())
```

### Erlang

Utilizada en simulación de colas, tiempor para completar una tarea ...

Acepta dos parámetros, media y k, que es un parámetro que indica cómo de picuda es la distribución

```python
from scipy.stats import erlang

# "picudez", k
# media
data = erlang.rvs(1, 10, size=1000)

sns.histplot(data=data, bins=100)

print(data.min(), data.max(), data.mean(), data.std())
```

```python
from scipy.stats import erlang

# media
# "picudez", k
data = erlang.rvs(15, 1, size=1000)

sns.histplot(data=data, bins=100)

print(data.min(), data.max(), data.mean(), data.std())
```

```python
from scipy.stats import erlang

# media
# "picudez", k
data = erlang.rvs(10, 100, size=1000)

sns.histplot(data=data, bins=100)

print(data.min(), data.max(), data.mean(), data.std())
```

### Weibull

Tiempos entre fallos, tiempo para completar una tarea

```python
from scipy.stats import weibull_min

# "picudez", k
# media
data = weibull_min.rvs(10, size=1000)

sns.histplot(data=data, bins=100)

print(data.min(), data.max(), data.mean(), data.std())
```

```python
from numpy import random as rn

data = rn.weibull(10, size=1000)
sns.histplot(data=data, bins=100)

print(data.min(), data.max(), data.mean(), data.std())
```

```python
from numpy import random as rn

data = rn.weibull(1, size=1000)
sns.histplot(data=data, bins=100)

print(data.min(), data.max(), data.mean(), data.std())
```

```python
from numpy import random as rn

data = rn.weibull(9, size=1000)
sns.histplot(data=data, bins=100)


print(data.min(), data.max(), data.mean(), data.std())
```
