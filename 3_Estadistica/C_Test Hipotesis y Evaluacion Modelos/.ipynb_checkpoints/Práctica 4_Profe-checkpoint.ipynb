
# Modelo Predictivo

Vamos a realizar un modelo de regresión lineal de dos variables.

En estos modelos es habitual llamar al conjunto de variables que actuarán como predictores X y a la variable a predecir y; existen otras convenciones como y e y con el acento circunflejo (^) pero, por facilidad al escribir en ventanas de código, utilizamos X e y.

En nuestro caso utilizaremos el número de visitas como predictor del gasto.

Tanto del diagrama de dispersión anterior como de la matriz de correlación, podemos ver que hay una dependencia muy fuerte entre ambas variables. 

Ahora se trata de obtener ese resultado de forma analítica y, además, conseguir la recta de regresión que nos permita extrapolar valores en el futuro.

Utilizamos el modelo de regresión por ser el más simple, no se entran en los detalles de estos modelos ya que habrá una clase específica para ellos.

```python
# Tenemos que utilizar .values.reshape(-1,1) porque el estimador lineal requiere un array de arrays

X = df['visitas'].values.reshape(-1,1)
y = df['gasto'].values.reshape(-1,1)
```

Para ajustar y comprobar la calidad del modelo utilizamos el mismo código que en la sesión de Feature Engineering.

El primer paso es crear un modelo "en blanco".

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

Ajustamos el modelo:

```python
lr.fit(X, y)
```

Comprobamos su calidad

```python
lr.score(X,y)
```

Una vez generado el modelo podemos consultar la pendiente de la recta (estamos en un modelo de regresión de una variable):

```python
lr.coef_
```

y el punto de corte con el eje y:

```python
lr.intercept_
```

Si os fijáis estos valores son en realidad arrays, ya que, de forma general, tendremos más de un multiplicador y más de un punto de corte (uno por dimensión).

Podemos utilizar estos valores para ver qué nos predice el modelo

```python
df['predicted'] = lr.predict(df['visitas'].values.reshape(-1,1))
df['predicted_manual'] = df['visitas'] * lr.coef_[0] + lr.intercept_[0]
df['error'] = df['predicted'] - df['gasto']
df.head()
```

Es habitual dibujar el diagrama de dispersión con la gráfica superpuesta:

```python
sns.scatterplot(data=df, x='visitas', y='gasto')
sns.lineplot(data=df, x='visitas', y='predicted', color='red')
```

## Resumen de errores

Una vez tenemos un modelo predictivo podemos ver su error:

Para ello utilizaremos el conjunto de métricas de sklearn



```python
from sklearn import metrics

print('Error medio absoluto:', metrics.mean_absolute_error(df['gasto'], df['predicted']))
print('Error medio cuadrado:', metrics.mean_squared_error(df['gasto'], df['predicted']))
print('Raíz del error medio cuadrado:', np.sqrt(metrics.mean_squared_error(df['gasto'], df['predicted'])))
```

```python
pl = sns.jointplot(data=df, x='visitas', y='error')
pl.ax_joint.plot([0,40], [0,0])
pl.fig.set_size_inches(14,6)
pl.fig.axes[0].set_xlim(1,39)
```

# Dataset COVID

```python
import pandas as pd
import seaborn as sns
pd.options.display.max_columns=100
from sklearn.linear_model import LinearRegression
from sklearn import metrics

def leer_csv_covid():
    columnas = ['stringency_index','population_density','median_age','aged_65_older',
          'aged_70_older','gdp_per_capita','extreme_poverty','cardiovasc_death_rate',
          'diabetes_prevalence','female_smokers','male_smokers', 'handwashing_facilities',
          'hospital_beds_per_thousand','life_expectancy','human_development_index']
    
    # Leemos el fichero
    df = pd.read_csv('./owid-covid-data.csv')
    
    df.drop(columns=['new_deaths','new_deaths_smoothed'], inplace=True)
    
    # Interpolamos
    df.loc[:,df.select_dtypes('number').columns.to_list()].interpolate(axis=1, inplace=True)
    
    # Última fila de cada país
    df = df.groupby(by='iso_code').nth(-1).reset_index(level=0)
    
    # Borramos columnas con pocos datos
    df.dropna(axis=1, thresh=50, inplace=True)
    
    # Borramos las columnas de fecha y país
    df.drop(columns =['location', 'date'], inplace=True)
    
    # Eliminamos las filas que no tienen información de fallecidos
    df = df.loc[pd.notna(df['total_deaths'])]

    # Imputamos la media
    medias = {c:df[c].mean() for c in columnas}
    df.fillna(medias, inplace=True)
    
    return df

df_covid = leer_csv_covid()
df_covid.drop(columns=['human_development_index'], inplace=True)
```
