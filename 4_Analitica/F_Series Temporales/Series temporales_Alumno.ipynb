# Introducción a series temporales

Una serie temporal es un conjunto de observaciones de una variable recogidas secuencialmente, es decir, se encuentran **en un orden específico**. Estas observaciones podrían ser, por ejemplo, el total de unidades vendidas en un supermercado cada mes, el número de personas que visita una página web cada día o el precio de alquiler de pisos durante un periodo determinado. 

Esta **dimensión temporal adicional** presente en las series temporales es una fuente de información muy útil que no se encuentra en otros tipos de datasets donde cada observación es independiente de las demás. Por ejemplo, si analizamos los salarios de los trabajadores de una empresa, es irrelevante si los ordenamos de mayor a menor salario o viceversa. 

Generalmente, las observaciones se recogen en instantes de tiempo equiespaciados pero, si por algún motivo esto no es posible, es necesario o bien utilizar únicamente los datos de instantes equiespaciados, o bien acumular las observaciones en periodos de tiempo para los que tengamos series completas.

En caso de trabajar con datos faltantes podemos recurrir a los métodos utilizados en sesiones anteriores para tratar de cubrir pequeños periodos de tiempo.

## Objetivo de una Serie Temporal
> El objetivo principal del tratamiento de series temporales es realizar **predicciones sobre el futuro** basadas en la información que poseemos en tiempos anteriores.

Cuando tratamos con series temporales, lo primero es fijarse en una serie de elementos característicos:

- **La tendencia**: es un cambio de los datos a largo plazo respecto a un nivel medio, y puede ser tanto ascendiente como descendiente. Es decir, la tendencia se trata del cambio a largo plazo de la media y suele ser un movimiento suave.

- **La estacionalidad o efecto estacional**: consiste en la influencia de periodos de tiempo específicos: anuales, mensuales, diarios, etc. Por ejemplo, la venta de un supermercado aumenta a principios de mes y disminuye a finales del mismo (periodicidad mensual) o el paro laboral suele aumentar en invierno y disminuir en verano (periodicidad anual). Este tipo de efectos suelen tener una fácil explicación e incluso se pueden eliminar de la serie **desestacionalizándola**.

- **La aletoriedad o presencia de outliers**: los outliers son medidas muy discordantes con la tendencia general. Una vez localizados los dos comportamientos anteriores (la tendencia y la estacionalidad), persistirán algunos componentes aleatorios que nos interesará caracterizar mediante algún tipo de modelo probabilístico que nos permitirá conocer el comportamiento a largo plazo de la serie. 
      
     

## Clasificación de las series temporales

Las series temporales suelen clasificarse en dos grandes grupos:

- **Estacionarias**: Las series estacionarias se caracterizan por una media y variabilidad constantes a lo largo del tiempo. Es decir, las propiedades de la serie no dependen del momento en el que se observan y sus valores oscilan alrededor de un valor medio constante, siempre con una amplitud también constante en el tiempo. Por lo tanto, las series que presentan una tendencia o estacionalidad, no serán estacionarias. Sin embargo, notar que sí pueden tener comportamientos cíclicos de periodo no conocido. Además, otra característica importante de estas series es que están **muy poco autocorreladas**, es decir, el valor de una observación concreta tiene poca relación con los valores anteriores, existe aleatoriedad.


- **No estacionarias**: Las series no estacionarias se caracterizan por una media y variabilidad que cambian a lo largo del tiempo. Estos cambios en la media son los que determinan la tendencia a crecer o decrecer a lo largo del tiempo. Además, están **altamente autocorreladas**, lo que quiere decir que cada observación de la serie temporal depende en gran medida de alguna observación pasada.

## Tests de estacionaridad

Como veremos más adelante, en general, para aplicar cualquier modelo a una serie temporal es necesario que ésta sea estacionaria. En el caso de que no lo sea, necesitaremos extraer la tendencia y estacionalidad antes de aplicar cualquier modelo de predicción. 

Para ello, lo primero será comprobar si la serie cumple las características de estacionariedad. Existen varios métodos: 
* Calcular la **media móvil** y la **varianza móvil** y ver si se mantienen constantes con el tiempo. Esta es una técnica visual que no nos permite saber con certeza si la serie es estacionaria o no, pero nos permite tener una idea más clara. 
* Dividir el dataset en subsets más pequeños y calcular la media y la varianza en cada uno de ellos, y ver si hay diferencias significativas. 
* Por último, existen test estadísticos diseñados específicamente para comprobar si una serie temporal es estacionaria. De entre los tests estadísticos, uno de los test más utilizados es **el test de Dickey–Fuller aumentado (ADF)**.

### Plots de la serie temporal

Lo primero que debemos hacer al trabajar con series temporales, es dibujarlas.

Realizar un plot de una serie temporal es el equivalente a realizar un histograma de un atributo; en series temporales este análisis visual es aún más importante ya que lo que queremos es ver cómo varía una variable a lo largo del tiempo.

Utilizaremos dos series temporales diferentes para mostrar y, para cada una de ellas calcularemos la media y varianza móvil.

Estas medidas se calculan con funciones de **pandas** de ventana (window), estas funciones trabajan de forma similar a los PARTITION BY que escribíamos en SQL, cada vez que llamamos a una función de ventana nos quedamos solo con las últimas **n** observaciones y, el agregado que obtenemos, se refiere a ellas.

Un ejemplo sería:

> df['ATRIBUTO'].rolling(window = 5).mean()

Que realiza la media móvil de tamaño 5.

Utilizando las series de datos de nacimientos y pasajeros que ya vimos antes obtenemos:

````python
df_births['BIRTHS_ROLLING'] = df_births['BIRTHS'].rolling(window = 10).mean()
df_births['BIRTHS_ROLLING_STD'] = df_births['BIRTHS'].rolling(window = 10).std()

df_passengers['PASSENGERS_ROLLING'] = df_passengers['PASSENGERS'].rolling(window = 10).mean()
df_passengers['PASSENGERS_ROLLING_STD'] = df_passengers['PASSENGERS'].rolling(window = 10).std()
````

````python
df_births.head(12)
````

````python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize = (20, 25))

ax1.plot(df_births['DATE'], df_births['BIRTHS'], color = 'black', label = 'Births', alpha = 0.4)
ax1.plot(df_births['DATE'], df_births['BIRTHS_ROLLING'], color = 'red', label = 'Moving average')
ax1.plot(df_births['DATE'], df_births['BIRTHS_ROLLING'] + df_births['BIRTHS_ROLLING_STD'], color = 'green', label = 'Moving variance')
ax1.plot(df_births['DATE'], df_births['BIRTHS_ROLLING'] - df_births['BIRTHS_ROLLING_STD'], color = 'green', label = 'Moving variance')

ax1.set_ylabel('Births')
ax1.legend()

ax2.plot(df_passengers['DATE'], df_passengers['PASSENGERS'], color = 'black', label = 'Passengers', alpha = 0.4)
ax2.plot(df_passengers['DATE'], df_passengers['PASSENGERS_ROLLING'], color = 'red', label = 'Moving average')
ax2.plot(df_passengers['DATE'], df_passengers['PASSENGERS_ROLLING'] + df_passengers['PASSENGERS_ROLLING_STD'], color = 'green', label = 'Moving variance')
ax2.plot(df_passengers['DATE'], df_passengers['PASSENGERS_ROLLING'] - df_passengers['PASSENGERS_ROLLING_STD'], color = 'green', label = 'Moving variance')
ax2.set_ylabel('Passengers')
ax2.legend()
````

En una serie temporal estacionaria, la media móvil y la desviación móvil seguirán una línea (más o menos) recta, fluctuando un poco.

Además, en una serie estacionaria, el histograma de sus valores seguirá una distribución aproximadamente normal.

Algunas series no estacionarias pueden convertirse en estacionarias con métodos que veremos más adelante.

````python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (25, 13))

ax1.hist(df_births['BIRTHS'], bins = 40, color = 'mediumblue')

ax2.hist(df_passengers['PASSENGERS'], bins = 40, color = 'green')
````

### Resumen de estadísticas

El método de prueba de la estacionaridad mediante el resumen de estadísicas consiste en dividir el dataset en grupos más o menos iguales y comprobar si la media y la varianza se mantienen constantes entre cada uno de ellos.

Esto nos dará una idea intuitiva de si algunos datasets son estacionarios o no. En el caso del dataset de los pasajeros, es evidente que no se trata de una serie temporal estacionaria. 

Sin embargo, en el caso del dataset de los nacimientos podríamos tener dudas sobre si lo es o no, dado que aunque las diferencias parecen muy pequeñas, la media y varianza no son constantes. 

````python
births1 = df_births[0:120]
births2 = df_births[120:240]
births3 = df_births[240:366]
````

````python
print('Grupo 1: ', births1['BIRTHS'].mean(), births1['BIRTHS'].std())
print('Grupo 2: ', births2['BIRTHS'].mean(), births2['BIRTHS'].std())
print('Grupo 3: ', births3['BIRTHS'].mean(), births3['BIRTHS'].std())
````

````python
pass1 = df_passengers[0:48]
pass2 = df_passengers[48:96]
pass3 = df_passengers[96:144]
````

````python
print('Grupo 1: ', pass1['PASSENGERS'].mean(), pass1['PASSENGERS'].std())
print('Grupo 2: ', pass2['PASSENGERS'].mean(), pass2['PASSENGERS'].std())
print('Grupo 3: ', pass3['PASSENGERS'].mean(), pass3['PASSENGERS'].std())
````

````python
def split_dataset(data,col,sets=2):
    splits = int(round(data.shape[0]/sets,0))
    for k,i in zip(range(1,sets+1),range(0,data.shape[0],splits)):
        subs = data[i:i+splits]
        print('Grupo {} (desde {} a {}): '.format(k,i,i+splits), subs[col].mean(), subs[col].std())

print(split_dataset(df_births,'BIRTHS',5))
split_dataset(df_passengers,'PASSENGERS',5)
````

### Test estadísticos: Test ADF

La manera más fiable de comprobar si una serie temporal es estacionaria o no es aplicar un test estadístico. Uno de los más utilizados es el test ADF.

Este test realiza una hipótesis de partida basada en características intrínsecas de los modelos estacionarios.

La hipótesis nula que realiza es la existencia de una raíz unitaria. Sin entrar en detalles, la existencia de esta raíz es una característica de un sistema no estacionario.

Por lo tanto lo que queremos es saber con qué seguridad podemos rechazar la hipótesis nula.

El test nos devuelve tanto el valor resultante (cuanto más negativo más probable es que la serie sea estacionaria) como el p-valor y diferentes valores críticos.

En python podemos implementarlo mediante un método de la librería statsmodels [statsmodels](https://www.statsmodels.org/stable/index.html), llamado [adfuller](https://www.statsmodels.org/stable/generated/statsmodels.tsa.stattools.adfuller.html).

````python
from statsmodels.tsa.stattools import adfuller

birth_result = adfuller(df_births['BIRTHS'].values)

print(f'Valor ADF: {birth_result[0]:.2f}')
print(f'p-valor: {birth_result[1]:.4f}')
print(f'Valores críticos:')
for key, valor in birth_result[4].items():
    print(f'\t{key} % :{valor:.2f}')
````

````python
passengers_result = adfuller(df_passengers['PASSENGERS'].values)

print(f'Valor ADF: {passengers_result[0]:.2f}')
print(f'p-valor: {passengers_result[1]:.4f}')
print(f'Valores críticos:')
for key, valor in passengers_result[4].items():
    print(f'\t{key} % :{valor:.2f}')
````

## Análisis de series temporales vs. regresión

La diferencia más significativa entre los problemas típicos de regresión y las series temporales es que estas últimas consideran la **autocorrelación**, es decir, la dependencia de un valor de la serie temporal con los valores anteriores. En una regresión normal, se hace la suposición de que cada observación es independendiente del resto, o que al menos existe una correlación muy pequeña entre ellas. 

### Gráficos de autocorrelación
Para comprobar si una serie temporal está o no altamente auto-correlada se pueden utilizar lo que se conoce como **gráficos de autocorrelación**. En ellos podemos ver en el eje x el valor de auto-correlación obtenido para cada uno de los diferentes "lags". Por "lag" se entiende una cantidad de tiempo fija entre cada observación, por ejemplo, lag 1 hace referencia a la observación siguiente a una dada. En el eje y vemos el valor de la auto-correlación, que puede ir entre -1 y +1:

- Autocorrelación +1: Indica una fuerte correlación entre un valor de la serie temporal y los valores pasados/futuros.
- Autocorrelación 0: Indica una correlación débil entre los valores de la serie temporal.
- Autocorrelación -1: Indica una anti-correlación fuerte entre los valores de la serie temporal.

Pandas dispone de un método de dibujo de los grafos de autocorrelación integrado.



````python
pd.plotting.autocorrelation_plot(df_passengers['PASSENGERS'])
````

Como podemos ver en el gráfico anterior, para la serie temporal del número de viajeros en avión, cada dato de la serie temporal está altamente correlacionado con los datos inmediatamente anteriores (aproximadamente hasta las 5 anteriores), lo cual está de acuerdo con el hecho de que la serie es no estacionaria. 

````python
pd.plotting.autocorrelation_plot(df_births['BIRTHS'])
````

Sin embargo, en el caso del número de nacimientos, vemos que en general los datos no están correlados, como sucede en las series estacionarias. 

### Lag plots

También existe otro tipo de gráficos que nos ayudan a conocer si nuestros datos están autocorrelados o no, y por tanto a identificar patrones de comportamiento en ellos: los **lag plots**. En ellos se representan los datos en un determinado instante de tiempo contra los datos en otro instante de tiempo, habitualmente el lag 1. En función del resultado, tendremos distintas posibilidades:

- Si los puntos se distribuyen a lo largo de la diagonal que va desde abajo a la izquierda hasta arriba a la derecha, tendremos una **relación de correlación positiva**. 
- Si los puntos se distribuyen a lo largo de la diagonal que va desde arriba a la izquierda hasta abajo a la derecha, tendremos una **relación de correlación negativa**. 
- Si los puntos se distribuyen aleatoriamente sin formar ningún tipo de patrón, los datos están poco correlados. 
- Si los puntos se distribuyen formando algún otro tipo de patrón como círculos o elipses, sugiere que los datos están correlados pero con algún tipo de correlación distinta a la lineal.

Para comparar el funcionamiento de todos los métodos que vamos a ver vamos a utilizar el **error absoluto**, es decir, la suma de la diferencia en valor absoluto entre los valores predichos y los valores reales. 


Es importante señalar que existen múltiples métodos para medir el error, la ventaja de utilizar el absoluto es que simplifica la comparación. Otros métodos para medir el error de las predicciones son:

$$
\text{Absolute Error}: \sum{\left| \hat{y} - y \right|}
$$

$$
\text{Mean Absolute Error}: \frac{\sum{\left| \hat{y} - y \right|}}{n}
$$


$$ 
 \text{Sesgo}: \frac{\sum{\hat{y} - y}}{n}
$$

$$
\text{MSE - Mean Square Error}: \frac{\sum{(\hat{y} - y})^2}{n}
$$

$$
\text{RMSE - Root of Mean Square Error}: \sqrt{\frac{\sum{(\hat{y} - y})^2}{n}}
$$

$$
\text{MAPE - Mean Absolute Percentage Error}: \frac{\sum{\frac{\hat{y} - y}{y}}}{n}
$$


$$
\text{SMAPE - Symmetric Mean Absolute Percentage Error}: \frac{\sum{\frac{\hat{y} - y}{y+ \hat{y}}}}{n}
$$

En estas fórmulas **n** representa el número de predicciones, **y** es el valor real e **ŷ** es el valor que nosotros hemos predecido

Además, como referencia vamos a compararlos con el modelo más básico que podemos utilizar para predecir valores futuros: **utilizar como predictor el último valor de la serie**.

### AR


El **método de autoregressión** (Autoregression, AR) es el modelo más simple de todos los que vamos a ver, y supone que cada observación es una función lineal de las observaciones anteriores más un término estocástico. 

Dependiendo de cuántos términos *p* se consideren, el modelo se denomina método **AR(p)** (método AR de orden p). 

Este valor p es simplemente el número de observaciones hacia el pasado que vamos a utilizar para predecir la siguiente.

Este método necesita que se haya eliminado la tendencia y la estacionalidad de la serie temporal. 




````python
descomposition_passengers.resid
````

````python
descomposition_passengers.resid.loc[:'1959']
````

````python
from statsmodels.tsa.ar_model import AutoReg

# Utilizamos 3 datos anteriores para predecir el siguiente
# Le pasamos el valor del residuo (lo que queda después de quitarle la tendencia y desestacionalizar la serie)
model = AutoReg(descomposition_passengers.resid.loc[:'1959'].values, lags = 3, old_names = False)

# Entrenamos el modelo
model_fit = model.fit()

# Predecimos tantas observaciones como queramos
n_obs = 11
yhat_ar_passengers = model_fit.predict(start = len(descomposition_passengers.resid.loc[:'1959']), 
                         end = len(descomposition_passengers.resid.loc[:'1959']) + n_obs)

error_ar_passengers = abs((descomposition_passengers.resid.loc['1960'].values-yhat_ar_passengers)).sum()
error_medio_passengers = abs((descomposition_passengers.resid.loc['1960'].values-yhat_ar_passengers)).mean()

print("El error medio es: ", error_medio_passengers)
print("El error total es: ", error_ar_passengers)
print("Coefs del modelo:", model_fit.params)
````

Vemos que, para el ejemplo de los pasajeros, el error ha disminuido hasta 383 (antes era 912); podemos utilizar otro valor para los lags.

Los parámetros del modelo nos indican qué coeficientes tiene la regresión para estos valores, podemos ver que, en este caso el lag ante-anterior a nuestro valor es el que tiene mayor peso en la predicción.

````python
# Utilizamos 10 datos anteriores para predecir el siguiente
model = AutoReg(descomposition_passengers.resid.loc[:'1959'].values, lags = 10, old_names = False)

# Entrenamos el modelo
model_fit = model.fit()

# Predecimos tantas observaciones como queramos
yhat_ar_passengers = model_fit.predict(start = len(descomposition_passengers.resid.loc[:'1959']), 
                         end = len(descomposition_passengers.resid.loc[:'1959']) + n_obs)

error_ar_passengers = abs((descomposition_passengers.resid.loc['1960'].values-yhat_ar_passengers)).sum()

print("El error total es: ", error_ar_passengers)
print("Parámetros del modelo:", model_fit.params)
````

Aunque el error ha disminuido, este no parece un método muy eficiente de encontrar el parámetro adecuado.

Al utilizar 10 lags, el peso cambia y parece que todos tienen influencia en nuestra medida.

Statsmodels nos proporciona un método para escoger el mejor orden a la hora de realizar la autoregresión.

````python
from statsmodels.tsa.ar_model import AutoReg, ar_select_order

lags = ar_select_order(descomposition_passengers.resid.loc[:'1959'].values, maxlag = 20, old_names = False)

model = AutoReg(descomposition_passengers.resid.loc[:'1959'].values, lags = lags.ar_lags, old_names = False)
model_fit = model.fit()

yhat_ar_passengers = model_fit.predict(start = len(descomposition_passengers.resid.loc[:'1959']), 
                         end = len(descomposition_passengers.resid.loc[:'1959']) + n_obs)

error_ar_passengers = abs((descomposition_passengers.resid.loc['1960'].values-yhat_ar_passengers)).sum()
print("El error total es: ", error_ar_passengers)
print("El número adecuado de 'lags' es: ", lags.ar_lags)
print("Parámetros del modelo:", model_fit.params)

````

### ARIMA

Este modelo de autorregresión integrada con media móvil (Autoregressive Integrated Moving Average, ARIMA) es una combinación de los dos anteriores que combina la modelización de las observaciones posteriores como una función lineal de las observaciones diferenciadas y los errores residuales para tiempos anteriores. 

El paso de diferenciación que hace que la serie temporal sea estacionaria se conoce como integración (I). Para especificar el orden de este modelo se dan por separado los órdenes de cada uno de los "submodelos": ARIMA(p, d, q).

````python
model = ARIMA(descomposition_passengers.resid.loc[:'1959'].values, order = (1, 1, 1))
model_fit = model.fit()

yhat_arima_passengers = model_fit.predict(start = len(descomposition_passengers.resid.loc[:'1959'].values), 
                         end = len(descomposition_passengers.resid.loc[:'1959'].values) + n_obs)

error_arima_passengers = abs((descomposition_passengers.resid.loc['1960'].values-yhat_arima_passengers)).sum()
print("El error total es: ", error_arima_passengers)
````

# Predicción de series temporales con Prophet

Prophet es una librería de código abierto desarrollada por Facebook que **permite realizar predicciones con series temporales de una sola variable**. Una de las grandes ventajas que incorpora esta librería es la posibilidad de tratar de una manera muy sencilla **eventos como vacaciones y festivos**, además de trabajar de forma más natural la tendencia y estacionalidad de la serie. 

Prophet es una librería que está optimizada para realizar predicciones sobre series temporales en el **ámbito de negocio**. Las características de este tipo de series temporales suelen ser algunas de las siguientes:

- Observaciones de la variable de análisis en periodos de horas, días o semanas, típicamente durante periodos de al menos un año. 
- Estacionalidades ligadas al comportamiento humano, como por ejemplo el día de la semana o la época del año. 
- Series temporales con vacaciones o  festivos importantes que ocurren a intervalos irregulares que se conocen con antelación, como por ejemplo el Viernes Santo.
- Cambios de tendencia históricos, por ejemplo debidos al lanzamiento de productos.
- Observaciones faltantes u outliers importantes. 
- Tendencias con curvas de crecimiento no lineales. 

## Funcionamiento de Prophet

El funcionamiento de Prophet se basa en lo que se conoce como un modelo de regresión aditiva. Un modelo de regresión aditiva asume que la serie temporal puede describirse como la suma de cuatro componentes independientes: 

- Para **la tendencia** utiliza una función a trozos con un crecimiento lineal o logístico. De esta manera, Prophet detecta automáticamente cambios en la tendencia seleccionando los puntos de cambio en los datos. 
- Una **componente estacional anual** modelada utilizando series de Fourier. 
- Una **componente estacional semanal** utilanzo variables dummy.
- Una **lista de festivos** importantes que proporciona el propio usuario.

Antes hemos visto que en algunos casos podemos contar con datos duplicados o con más de una observación para cada instante. Sin embargo, también podría suceder que contemos con observaciones equiespaciadas (o no) pero con una frecuencia que no nos interesa. Por ejemplo, si tenemos observaciones cada 2 horas pero nos gustaría tener una única observación por día, podemos simplemente calcularla como la media de todas las observaciones de cada día:

````python
data_index = pd.date_range(start = '01/01/2021', periods = 500, freq = '2H')
````

````python
data = pd.Series(np.arange(500), index = data_index)
data
````

````python
data.groupby(pd.Grouper(freq = 'D')).mean().head(10)
````

Una opción, como acabamos de ver, es hacer un groupby y calcular la media del periodo, sin embargo pandas ya tiene métodos específicos para hacer resampling: [resample](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.resample.html). En concreto, se conoce como resampling al proceso de convertir una serie de tiempo de una frecuencia determinada a otra distinta. 

````python
rng = pd.date_range('2000-01-01', periods = 100, freq = 'D')
t = pd.Series(np.random.randn(len(rng)), index = rng)
t
````

````python
t.resample('M').mean()
````

````python
t.resample('M', kind = 'period').mean()
````

Cuando se agregan datos de frecuencias altas a otras frecuencias regulares más bajas se denomina **downsampling**. Los datos no necesitan tener una frecuencia concreta para poder agregarlos, dado que se definen una serie de intervalos y se comprueba en cual de ellos cae cada una de las observaciones. Estos intervalos serán abiertos por un extremo y cerrados por el otro, y hay que definir cual de los extremos será el cerrado. También es necesario pensar cómo se van a etiquetar los intervalos, si con el principio o con el final. 

````python
rng = pd.date_range('2000-01-01', periods = 21, freq = 'T')
t = pd.Series(np.arange(21), index = rng)
t
````

````python
t.resample('5min', closed = 'right', label = 'right').sum()
````

Otra tarea que nos podemos encontrar con frecuencia es el **upsampling**, cuando queremos transformar nuestros datos de una frecuencia más baja a otra más alta. En este caso no es necesario realizar ningún tipo de agregación, pero como veremos se introducirán valores faltantes:

````python
rng = pd.date_range('2014-06-24', periods = 5, freq = 'T')
df = pd.DataFrame({'Value': np.random.randn(len(rng))}, index = rng) 
df
````

````python
df.resample('20S').backfill(limit = 1).head(6)
````

````python
df.resample('20S').ffill(limit = 1).head(6)
````

````python
df.resample('10S').nearest(limit = 2).head(10)
````

## Moving window functions

Otra clase de operaciones que nos pueden interesar es aplicar funciones sobre ventanas temporales de nuestros datos. Un ejemplo es la media móvil que vimos al principio, y que se calculaba simplemente como:

````python
df.rolling(window = 2).mean()
````

Pero podríamos aplicar cualquier función que deseemos después del rolling: las típicas son mean(), std(), median(), sum(), count(), max(), min(), ... Y también podemos aplicar varias funciones en un solo paso:

````python
df.rolling(window = 2, min_periods = 0).agg({"sum", "count", "mean"})
````
