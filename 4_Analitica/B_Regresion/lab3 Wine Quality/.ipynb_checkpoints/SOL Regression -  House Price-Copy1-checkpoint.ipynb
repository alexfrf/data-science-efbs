## 1. Load both data set: white-wine and red-whine, and join it (use concat) in a unique dataframe: df_wine. Answer all the follwing questions using the complete df_wine dataframe

```python
import pandas as pd
df_red = pd.read_csv("winequality-red.csv", sep=";")
print(df_red.shape)
df_red.info()
```

```python
df_white = pd.read_csv("winequality-white.csv", sep=";")
print(df_white.shape)
df_white.info()
```

```python
df_wine = pd.concat([df_red, df_white], axis=0)
df_wine.shape
```

## 2.  Missing inputation (in case there exists missing values)


```python
# check for missings
df_wine.info()
```

We can see, that columns "chlorides" and "free_sulfur_dioxide" have 5 miising values. Let's solve this issue

```python
df_wine.loc[df_wine["chlorides"].isnull(),"chlorides"] = df_wine["chlorides"].mean()
df_wine.loc[df_wine["free_sulfur_dioxide"].isnull(),"free_sulfur_dioxide"] = df_wine["free_sulfur_dioxide"].mean()


df_wine.info()
```

## 3. Exploratory Data Analysis. Show correlation and main statistics of all the variables in the data set

## 4. Categorical variables binning (dummies) - in case you have categorical variables


The variable color is categorical. So make dummies for it

```python
df_wine.info()
```

```python
# Creating dummy variables:
df_wine_dummies = pd.get_dummies(df_wine, columns=['color'])
df_wine_dummies.head()
```

```python
df_wine_dummies.columns
```

## 5. Split in train and test (25%)


```python
from sklearn.model_selection import train_test_split

# dataframe with the explicative variables
X = df_wine_dummies[ ['fixed_acidity', 'volatile_acidity', 'citric_acid', 'residual_sugar',
       'chlorides', 'free_sulfur_dioxide', 'total_sulfur_dioxide', 'density',
       'pH', 'sulphates', 'alcohol', 'color_red', 'color_white'] ]
y = df_wine_dummies['quality']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=1)
```

## 6. Use SQL to get the following statistics: the total number of bottles and average of alcohol grouped by the quality


```python
from pandasql import sqldf
q = """ select quality, count(*) as n, avg(alcohol) as mean_alcohol 
from df_wine
group by quality"""
df_result = sqldf( q, locals())

df_result.head()
```

## 7. Fit models to the train dataset. Models to use: linear regression and random forest


### Linear Regression

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

model_lr = lr.fit(X = X_train, y=y_train)
```

### Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100, max_depth=5)

model_rf = rf.fit(X = X_train, y=y_train)
```

## 7. Perfomance metrics to compute: MAE and relative MAE

```python
# 1. make forecast
y_test_predict_lr = lr.predict(X_test)
y_test_predict_rf = rf.predict(X_test)
```

```python
# 2. compute performance errors
from sklearn.metrics import mean_absolute_error

MAE_lr = mean_absolute_error(y_pred = y_test_predict_lr, y_true = y_test)
MAE_rf = mean_absolute_error(y_pred = y_test_predict_rf, y_true = y_test)

print("Linear Regression ------------------------")
print("MAE: ", MAE_lr)
print("Relative MAE: ", MAE_lr/y_test.mean() )

print(" ")
print("Random Forest Regression -----------------")
print("MAE: ", MAE_rf)
print("Relative MAE: ", MAE_rf/y_test.mean() )

```
