#  Business Problem

Given a  dataset of house prices in the USA. Build a model, to predict the price of a house base on the other variables. The dataset "houses.csv" contains the following variables

- 'Avg. Area Income': Avg. Income of residents of the city house is located in.
- 'Avg. Area House Age': Avg Age of Houses in same city
- 'Avg. Area Number of Rooms': Avg Number of Rooms for Houses in same city
- 'Avg. Area Number of Bedrooms': Avg Number of Bedrooms for Houses in same city
- 'Area Population': Population of city house is located in
- 'Price': Price that the house sold at
- 'Address': Address for the house


The target variable is "Price"; and the predictive variables are: 'Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms', 'Avg. Area Number of Bedrooms' and 'Area Population'


Your analysis must include:

1. Missing inputation (in case there exists missing values) 
2. Exploratory Data Analysis. Show scatterplots, correlation and main statistics of all the variables in the data set
3. Categorical variables binning (dummies) - in case you have categorical variables
4. Split in train and test (25%)
5. Use SQL to get the following statistics: the total number of houses and average of price
6. Fit models to the train dataset. Models to use: linear  regression and random forest
7. Perfomance metrics to compute: MAE and relative MAE



## 0. Import the data


```python
import pandas as pd
df_housing = pd.read_csv("housing.csv")
df_housing.head()
```

## 1.  Missing inputation (in case there exists missing values)


```python
# check for missings
df_housing.info()
```

We can see, that column "Avg. Area Income" has 4.997 non-null values, so 3 miising values. Let's solve this issue

```python
df_housing.loc[df_housing["Avg. Area Income"].isnull(),"Avg. Area Income"] = df_housing["Avg. Area Income"].mean()


df_housing.info()
```

## 2. Exploratory Data Analysis. Show scatterplots, correlation and main statistics of all the variables in the data set

## 3. Categorical variables binning (dummies) - in case you have categorical variables

## 4. Split in train and test (25%)


```python
from sklearn.model_selection import train_test_split

# dataframe with the explicative variables
X = df_housing[ ['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population'] ]
y = df_housing['Price']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=1)
```

## 5. Use SQL to get the following statistics: the total number of houses and average of price


```python
from pandasql import sqldf

df_result = sqldf( "select count(*) as tot_houses, avg(Price) as mean_price from df_housing", locals())

df_result.head()
```

## 6. Fit models to the train dataset. Models to use: linear regression and random forest


### Linear Regression

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

model_lr = lr.fit(X = X_train, y=y_train)
```

### Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100, max_depth=5)

model_rf = rf.fit(X = X_train, y=y_train)
```

## 7. Perfomance metrics to compute: MAE and relative MAE

```python
# 1. make forecast
y_test_predict_lr = lr.predict(X_test)
y_test_predict_rf = rf.predict(X_test)
```

```python
# 2. compute performance errors
from sklearn.metrics import mean_absolute_error

MAE_lr = mean_absolute_error(y_pred = y_test_predict_lr, y_true = y_test)
MAE_rf = mean_absolute_error(y_pred = y_test_predict_rf, y_true = y_test)

print("Linear Regression ------------------------")
print("MAE: ", MAE_lr)
print("Relative MAE: ", MAE_lr/y_test.mean() )

print(" ")
print("Random Forest Regression -----------------")
print("MAE: ", MAE_rf)
print("Relative MAE: ", MAE_rf/y_test.mean() )

```
