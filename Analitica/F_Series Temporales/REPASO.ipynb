## REPASO - Series Temporales

### Clasificación de las series temporales // Estacionalidad
Las series temporales suelen clasificarse en dos grandes grupos:

- **Estacionarias**: Las series estacionarias se caracterizan por una media y variabilidad constantes a lo largo del tiempo. Es decir, las propiedades de la serie no dependen del momento en el que se observan y sus valores oscilan alrededor de un valor medio constante, siempre con una amplitud también constante en el tiempo. Por lo tanto, las series que presentan una tendencia o estacionalidad, no serán estacionarias. Sin embargo, notar que sí pueden tener comportamientos cíclicos de periodo no conocido. Además, otra característica importante de estas series es que están **muy poco autocorreladas**, es decir, el valor de una observación concreta tiene poca relación con los valores anteriores, existe aleatoriedad.


- **No estacionarias**: Las series no estacionarias se caracterizan por una media y variabilidad que cambian a lo largo del tiempo. Estos cambios en la media son los que determinan la tendencia a crecer o decrecer a lo largo del tiempo. Además, están **altamente autocorreladas**, lo que quiere decir que cada observación de la serie temporal depende en gran medida de alguna observación pasada. 


Para poder aplicar series temporales es necesario que los datos presenten un comportamiento estacionario. En el caso de que no lo sea, necesitaremos extraer la tendencia y estacionalidad antes de aplicar cualquier modelo de predicción.

Para ello, lo primero será comprobar si la serie cumple las características de estacionariedad. Existen varios métodos:

- Calcular la media móvil y la varianza móvil y ver si se mantienen constantes con el tiempo. Esta es una técnica visual que no nos permite saber con certeza si la serie es estacionaria o no, pero nos permite tener una idea más clara.
- Dividir el dataset en subsets más pequeños y calcular la media y la varianza en cada uno de ellos, y ver si hay diferencias significativas.
- Por último, existen test estadísticos diseñados específicamente para comprobar si una serie temporal es estacionaria. De entre los tests estadísticos, uno de los test más utilizados es el test de Dickey–Fuller aumentado (ADF).

Igualmente, lo primero que tendremos que hacer al encontrarnos con una serie temporal es representarla gráficamente.

```python
def split_dataset(data,col,sets=2):
    splits = int(round(data.shape[0]/sets,0))
    for k,i in zip(range(1,sets+1),range(0,data.shape[0],splits)):
        subs = data[i:i+splits]
        print('Grupo {} (desde {} a {}): '.format(k,i,i+splits), subs[col].mean(), subs[col].std())

print(split_dataset(df_births,'BIRTHS',5))
split_dataset(df_passengers,'PASSENGERS',5)
```

#### 3- Tests Estadísticos de Estacionalidad

**Test ADF**

H0: La serie es estacionaria

```python
from statsmodels.tsa.stattools import adfuller

birth_test = adfuller(df_births.BIRTHS.values)
print(f'Valor ADF: {birth_test[0]:.2f}')
print(f'p-valor: {birth_test[1]:.4f}')
print(f'Valores críticos:')
for key, valor in birth_test[4].items():
    print(f'\t{key} % :{valor:.2f}')
```

PValor es < 0.05, así que no rechazamos la hipótesis nula de que Births sea estacionario.

```python
pass_test = adfuller(df_passengers.PASSENGERS.values)
print(f'Valor ADF: {pass_test[0]:.2f}')
print(f'PValue: {pass_test[1]:.2f}')
print(f'Valores Críticos:')
for key,value in pass_test[4].items():
    print(key,':',f'{value:.2f}')
```

PValor es > 0.05, así que rechazamos la hipótesis nula de que Births sea estacionario. Al ser no estacionaria, debemos desestacionalizar la serie

### Series Temporales vs Regresión // Autocorrelación

La diferencia más significativa entre los problemas típicos de regresión y las series temporales es que estas últimas consideran la autocorrelación, es decir, la dependencia de un valor de la serie temporal con los valores anteriores. En una regresión normal, se hace la suposición de que cada observación es independendiente del resto, o que al menos existe una correlación muy pequeña entre ellas.

**Cómo comprobar la Autocorrelación en Series**

#### 1- Gráficos de Autocorrelación

```python
pd.plotting.autocorrelation_plot(df_passengers['PASSENGERS'])
```

Como podemos ver en el gráfico anterior, para la serie temporal del número de viajeros en avión, cada dato de la serie temporal está altamente correlacionado con los datos inmediatamente anteriores (aproximadamente hasta las 5 anteriores), lo cual está de acuerdo con el hecho de que la serie es no estacionaria.

```python
pd.plotting.autocorrelation_plot(df_births.BIRTHS)
```

No existe autocorrelación apenas, lo que es lógico teniendo en cuenta que la serie de nacimientos es estacionaria

#### 2 - LagPlots

También existe otro tipo de gráficos que nos ayudan a conocer si nuestros datos están autocorrelados o no, y por tanto a identificar patrones de comportamiento en ellos: los lag plots. En ellos se representan los datos en un determinado instante de tiempo contra los datos en otro instante de tiempo, habitualmente el lag 1. En función del resultado, tendremos distintas posibilidades:

Si los puntos se distribuyen a lo largo de la diagonal que va desde abajo a la izquierda hasta arriba a la derecha, tendremos una relación de correlación positiva.
Si los puntos se distribuyen a lo largo de la diagonal que va desde arriba a la izquierda hasta abajo a la derecha, tendremos una relación de correlación negativa.
Si los puntos se distribuyen aleatoriamente sin formar ningún tipo de patrón, los datos están poco correlados.
Si los puntos se distribuyen formando algún otro tipo de patrón como círculos o elipses, sugiere que los datos están correlados pero con algún tipo de correlación distinta a la lineal.

```python
pd.plotting.lag_plot(df_passengers['PASSENGERS'])
sns.lineplot(x=[df_passengers.PASSENGERS.min(),df_passengers.PASSENGERS.max()], y=[df_passengers.PASSENGERS.min(),df_passengers.PASSENGERS.max()], color='green', linestyle='--')
```

Los métodos habituales para el estudio de series temporales son los de la familia XARIMA que incluye:

    - AR

Estos modelos para la predicción de series temporales se basan en la suposición de que existen relaciones lineales entre las observaciones y, habitualmente, requieren que se elimine la tendencia y la estacionalidad de la serie temporal para que funcionen.

### Autorregresión (AR)

El método de autoregressión (Autoregression, AR) es el modelo más simple de todos los que vamos a ver, y supone que cada observación es una función lineal de las observaciones anteriores más un término estocástico.

PARÁMETROS: lags (nº de observaciones hacia el pasado empleadas para predecir)

```python
from statsmodels.tsa.ar_model import AutoReg

y_passengers = passengers['Resid'].loc[:str(passengers['Resid'].index[-1].year-1)]
y_passengers.tail()
```

```python
model = AutoReg(y_passengers, lags = 3, old_names = False)
model_fit = model.fit()
model_fit.params
```

```python
n_obs = 11
y_passengers_pred = model_fit.predict(start = len(y_passengers),end = len(y_passengers)+n_obs)
error_ar_passengers = abs((passengers['Resid'].loc['1960'].values-y_passengers_pred)).sum()
error_medio_passengers = abs((passengers['Resid'].loc['1960'].values-y_passengers_pred)).mean()

print("El error medio es: ", error_medio_passengers)
print("El error total es: ", error_ar_passengers)
print("\nCoefs del modelo:", model_fit.params)
```

*Statsmodels nos proporciona un método para escoger el mejor lag a la hora de realizar la autoregresión.*

```python
sns.scatterplot(data=passengers.Resid)
sns.scatterplot(x = passengers.Resid.index[-12:], y = y_passengers_pred,label='Pred_3_lags')
sns.scatterplot(x = passengers.Resid.index[-12:], y = passengers['Resid'].iloc[-12:],label='Real',color='Navy')
```

```python
from statsmodels.tsa.ar_model import AutoReg, ar_select_order

lags = ar_select_order(y_passengers.values,maxlag=48, old_names = False)
model = AutoReg(y_passengers.values, lags = lags.ar_lags,old_names=False)
model_fit = model.fit()
model_fit.params
```

```python
y_passengers_pred = model_fit.predict(start=len(y_passengers),end=n_obs+len(y_passengers))
error_ar_passengers = abs((passengers['Resid'].loc['1960'].values-y_passengers_pred)).sum()

print("El error total es: ", error_ar_passengers)
print("El número adecuado de 'lags' es: ", lags.ar_lags[-1])
print("Parámetros del modelo:", model_fit.params)
```

#### Moving Average (MA)

El método de la media móvil (Moving Average, MA) consiste en modelar la siguiente observación en la serie temporal como una función lineal de los residuos que resultan de una regresión anterior.

El objetivo es, sabiendo que estamos cometiendo un error, tratar de calcular cuál será ese error para contrarrestarlo.

Al igual que en el caso anterior, la notación del modelo nos dice el orden, por ejemplo MA(1) es el modelo de media móvil de orden uno.

Al igual que el anterior, este modelo se aplica a series temporales de una sola variable y sin tendencia ni estacionalidad.

```python
from statsmodels.tsa.arima.model import ARIMA

n_obs = 11
model = ARIMA(descomposition_passengers.resid.loc[:'1959'].values, order = (0, 0, 1))
model_fit = model.fit()

yhat_ma_passengers = model_fit.predict(start = len(descomposition_passengers.resid.loc[:'1959'].values), 
                         end = len(descomposition_passengers.resid.loc[:'1959'].values) + n_obs)

error_ma_passengers = abs((descomposition_passengers.resid.loc['1960'].values-yhat_ma_passengers)).sum()

print("El error total es: ", error_ma_passengers)
print("Parámetros del modelo:", model_fit.params)
```

#### ARIMA

Este modelo de autorregresión integrada con media móvil (Autoregressive Integrated Moving Average, ARIMA) es una combinación de los dos anteriores que combina la modelización de las observaciones posteriores como una función lineal de las observaciones diferenciadas y los errores residuales para tiempos anteriores.

El paso de diferenciación que hace que la serie temporal sea estacionaria se conoce como integración (I). Para especificar el orden de este modelo se dan por separado los órdenes de cada uno de los "submodelos": ARIMA(p, d, q).

### Predicción en series temporales con PROPHET

Prophet permite realizar predicciones con series temporales de una sola variable. Una de las grandes ventajas que incorpora esta librería es la posibilidad de tratar de una manera muy sencilla eventos, además de trabajar de forma más natural la tendencia y estacionalidad de la serie.

EJEMPLOS

    - Observaciones de la variable de análisis en periodos de horas, días o semanas, típicamente durante periodos de al menos un año.
    - Estacionalidades ligadas al comportamiento humano, como por ejemplo el día de la semana o la época del año.
    - Series temporales con vacaciones o festivos importantes que ocurren a intervalos irregulares que se conocen con antelación, como por ejemplo el Viernes Santo.
    - Cambios de tendencia históricos, por ejemplo debidos al lanzamiento de productos.
    - Observaciones faltantes u outliers importantes.
    - Tendencias con curvas de crecimiento no lineales.

**P order**

The value of P can be extracted by looking at the Partial Autocorrelation (PACF) graph of the seasonal component. PACF can be imagined as the correlation between the series and its lag, after excluding the contributions from the intermediate lags.

```python
plot_pacf(seasonal, lags =15)
plt.show()
```

```python
P =15
```

**Q Order**

The Q order can be calculated from the Autocorrelation (ACF) plot. Autocorrelation is the correlation of a single time series with a lagged copy of itself.

```python
plot_acf(seasonal, lags =15)
plt.show()
```
