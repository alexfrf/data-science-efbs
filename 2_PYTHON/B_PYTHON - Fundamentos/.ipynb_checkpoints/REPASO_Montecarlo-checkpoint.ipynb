## Repaso: Simulación de Montecarlo y Cadenas de Markov

```python
import pandas as pd
import numpy as np
import random
```

```python
def invertir(dinero):
    if random.random() < 0.5:
        return dinero * 1.03
    else:
        return dinero * 0.97

def simulacion():
    dinero_inicial = 10000
    dinero = dinero_inicial
    for i in range(10000):
        dinero = invertir(dinero)
    
    return dinero

resultados_simulacion = []
for i in range(1000):
    resultados_simulacion.append(simulacion())
```

```python
print(sum(resultados_simulacion) / len(resultados_simulacion))
```

# Mi primera Cadena de Markov

Las cadenas de Markov se basan en la transición de estados.

Una transición de estados no es más que una tabla que identifica un estado de partida y varios estados de llegada, asignando una probabilidad diferente a cada estado de llegada según, habitualmente, datos históricos.

Si pensamos en una tienda que vende ropa podemos pensar en una transición de estados de una prenda de la siguiente forma.



| Ha tenido venta hoy | Tiene venta mañana             | Probabilidad |
| :----------------|:----------------------| :------|
| Si | Sí | 80%|
| Sí | No | 20% |
| No | Sí | 40% |
| No | No | 60% |

```python
def venta_manana(productos_venta_hoy, productos_sin_venta_hoy):
    productos_venta_manana = productos_venta_hoy * .8 + productos_sin_venta_hoy * .4
    productos_sin_venta_manana = productos_venta_hoy * .2 + productos_sin_venta_hoy * .6
    
    return productos_venta_manana, productos_sin_venta_manana
```

```python
venta_manana(3,1)
```

```python
productos_venta_hoy_inicial = 1000
productos_sin_venta_hoy_inicial = 0

htco_venta = [productos_venta_hoy_inicial]
htco_sin_venta = [productos_sin_venta_hoy_inicial]
productos_venta_manana = productos_venta_hoy_inicial
productos_sin_venta_manana = productos_sin_venta_hoy_inicial

for x in range(50):
    productos_venta_manana, productos_sin_venta_manana = venta_manana(productos_venta_manana, productos_sin_venta_manana)
    htco_venta.append(productos_venta_manana)
    htco_sin_venta.append(productos_sin_venta_manana)
```

## Fuentes y sumideros en Cadenas de Markov

Dentro de las cadenas de Markov, pueden existir sumideros (etados que no tienen ningún estado de salida que no sean ellos mismos) y fuentes (estados en lso que entran nuevos elementos en el sistema) aunque, en una cadena de Markov normal, no se introducen nuevos elementos con el tiempo.

Podemos pensar en una cadena de Markov un  poco más interesante si pensamos en un supermercado que tiene varias secciones:
 - Entrada
 - Carnicería
 - Pescadería
 - Bebidas
 - Frutería
 - Cajas
 - Salida
 
Con el siguiente diagrama de transiciones:


Suponemos que, una vez que entra en el supermercado tiene un 25% de posibilidades de ir a cualquier otra sección o a las cajas.

```python
posicion_actual = {
    'Entrada': 100,
    'Carniceria': 0,
    'Fruteria': 0,
    'Bebidas': 0,
    'Pesca': 0,
    'Cajas': 0,
    'Salida': 0
}


transicion = {
    'Entrada': {
       'Carniceria': .25,
        'Fruteria': .25,
        'Bebidas': .25,
        'Pesca': .25
    },
    'Carniceria': {
        'Fruteria': .25,
        'Bebidas': .25,
        'Pesca': .25,
        'Cajas': .25
    }, 
    'Fruteria': {
        'Carniceria': .25,
        'Bebidas': .25,
        'Pesca': .25,
        'Cajas': .25
    }, 
    'Bebidas': {
        'Carniceria': .25,
        'Fruteria': .25,
        'Pesca': .25,
        'Cajas': .25
    }, 
    'Pesca': {
        'Carniceria': .25,
        'Fruteria': .25,
        'Bebidas': .25,
        'Cajas': .25
    },
    'Cajas': {'Salida': 1},
    'Salida': {'Salida': 1}
}

def proxima_seccion(secciones_actuales):
    secciones_siguientes = {
        k:0 for k in secciones_actuales.keys()
    }
    for seccion in transicion.keys():
        for seccion_destino, probabilidad in transicion[seccion].items():
            secciones_siguientes[seccion_destino] += secciones_actuales[seccion] * probabilidad
    return secciones_siguientes



proxima_seccion(posicion_actual)
```

### Conversión a Monte Carlo

```python
situacion_inicial = posicion_actual = {
    'Entrada': 100,
    'Carniceria': 0,
    'Fruteria': 0,
    'Bebidas': 0,
    'Pesca': 0,
    'Cajas': 0,
    'Salida': 0
}


situacion_actual = situacion_inicial
for x in range(30):
    situacion_actual = proxima_seccion(situacion_actual)
    
print(situacion_actual)
```

### Monte Carlo con histórico

```python
situacion_inicial = {
    'Entrada': 100,
    'Carniceria': 0,
    'Fruteria': 0,
    'Bebidas': 0,
    'Pesca': 0,
    'Cajas': 0,
    'Salida': 0
}

htco = {
    k : [] for k in situacion_inicial.keys()
}


situacion_actual = situacion_inicial
for x in range(30):
    situacion_actual = proxima_seccion(situacion_actual)
    for k, v in situacion_actual.items():
        htco[k].append(v)
    
print(situacion_actual)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (16,12)

for k, v in htco.items():
    sns.lineplot(data=v)
    
```
