# Ejercicio Compendio

Lee el fichero 'FACT_VENTA.xlsx' y realiza las siguientes operaciones:


1. Obtén el margen real acumulado por TIENDA_ID
1. Obtén el margen acumulado por TIENDA_ID y MES_ID
1. Obtener la diferencia entre el margen teórico y el margen real por cliente
1. Multiplica por dos toda el Margen Teórico de Tienda 1
1. Filtra las ventas realizadas antes del 31 de Enero

```python
df = pd.read_excel('FACT_VENTA.xlsx')
df.head()
```

```python
print(df.columns)
df.shape
```

```python
margen_df = df.groupby(by=['TIENDA_ID','MES_ID'],as_index=False)['MARGEN_REAL'].mean().sort_values(by=['TIENDA_ID'],ascending=False)
margen_df
```

```python
margen_acum = df.groupby(by='TIENDA_ID',as_index=False)['MARGEN_REAL'].mean().sort_values(by='MARGEN_REAL',ascending=False)
margen_acum
```

```python
margen_acum = df.groupby(by='MES_ID',as_index=False)['MARGEN_REAL'].mean().sort_values(by='MES_ID',ascending=True)
margen_acum
```

```python
# Obtener la diferencia entre el margen teórico y el margen real por cliente

margenes = df.groupby(by='CLIENTE_ID',as_index=False)[['MARGEN_REAL','MARGEN_TEORICO']].mean()
margenes['DIF'] = margenes['MARGEN_TEORICO'] - margenes['MARGEN_REAL']
margenes.sort_values(by='DIF',ascending=False).head()
```

```python
# Multiplica por dos toda el Margen Teórico de Tienda 1
df_m = df[df['TIENDA_ID']=='Tienda 1']['MARGEN_TEORICO'] * 2
df_m.mean()
```

```python
df1 = df[df['MES_ID']<=201901]
df1_tienda = df1.groupby(by='TIENDA_ID',as_index=False)['MARGEN_REAL'].mean().sort_values(by='MARGEN_REAL',ascending=False)
df1_tienda
```
