 {

  }

## Consigue características comunes
Consigue un conjunto de columnas común en todos los dataframes (¿qué otras columnas además de País son comunes a todos los dataframes?).

Se trata de buscar columnas que tengan **el mismo nombre**, no que referencien el mismo concepto.

## Combinación
Combina todos los dataframes que contengan la columna ```País``` en uno solo con todas las columnas comunes.

## Filtrado 
Elimina todos los países cuya Superficie no sea un número.

## Adición de columnas
Crea dos nuevas columna que contengan:
 * La primera letra de cada país
 * La primera letra de cada capital


````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
mapping =  { 
   'Superficie (km²)': 'Superficie',
   'Superficie(km²)': 'Superficie'
   }
````

````python
import glob

ficheros = glob.glob('./*.xlsx', recursive=True)
lista_dfs = []
incl = []
for fichero in ficheros:
    df = pd.read_excel(fichero)
    df.rename(columns=mapping,inplace=True)
    cols = df.columns
    if 'País' in cols:
        lista_dfs.append(df)
        incl.append(int(fichero.replace('.\\datos_paises_','').replace('.xlsx','')))

df = pd.concat(lista_dfs, ignore_index=True)
combi = pd.concat(lista_dfs,join='inner',ignore_index=True)
````

````python
df.shape[0]
````

````python
combi.head()
````

````python
df = combi
````

````python
df.head()
````

````python
df['Superficie'] = df['Superficie'].str.replace('.','').replace(',','')
df = df[df['Superficie'].str.isdigit()==True]
df['Superficie'] = df['Superficie'].astype(int)
````

````python
df['pPaís'] = [j['País'][0] for i,j in df.iterrows()]
df['pCapital'] = [j['Capital'][0] for i,j in df.iterrows()]
````

````python
df['Quintil'] = pd.qcut(df['Superficie'], 5, labels = [1,2,3,4,5])
df
````

## Agregación
Obtén:

* La superficie total agregada por la inicial del país.
* El número de países por la inicial de su capital.
* El promedio de km2 para cada inicial de país.

````python
grouped1 = df.groupby(by='pPaís',as_index=False)['Superficie'].agg('sum').sort_values(by='Superficie',ascending=False)
grouped1
````

````python
grouped2 = df.groupby(by='pCapital',as_index=False)['Superficie'].agg('count').sort_values(by='Superficie',ascending=False)
grouped2
````

````python
grouped3 = df.groupby(by='pPaís',as_index=False)['Superficie'].agg('mean').sort_values(by='Superficie',ascending=False)
grouped3
````

## Visualización

Muestra:
 1. Superficie agregada por inicial como un gráfico de barras.
 2. Superficie como un histograma (sin agregar).
 3. Conteo de países por inicial de su capital como un gráfico de tarta.
 4. Gráfico de dispersión utilizando la inicial del país como eje X, la inicial de la capital como eje Y y la superficie como valor.

````python
plt.figure(figsize=(10,7))
plt.bar(x=grouped1['pPaís'],height=grouped1['Superficie'])
sns.despine()
````

````python
plt.figure(figsize=(10,7))
plt.hist(x=df['Superficie'],bins=10)
sns.despine();
````

````python
plt.figure(figsize=(12,8))
plt.pie(grouped2['Superficie'],labels=grouped2['pCapital'],autopct='%1.0f%%',shadow=True);
````

````python
plt.figure(figsize=(12,8))
plt.scatter(x=df['pPaís'],y=df['pCapital'],s=np.sqrt(df.Superficie)*0.75, c=df.Quintil, cmap='Reds')
sns.despine()
````
